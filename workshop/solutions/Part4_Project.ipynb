## Part 4 : Project for Practice -- A Memory Card Game!

Let's build a game called Memory Card Game. Player can beat the game by memorizing the card they flip and trying to match all 8 pairs with less moves!

![MemoryCard](../images/MemoryGame.gif)

Let's first import some libraries

```python
import random  # random allow us to generate random number
from IPython.display import HTML, display  # allow us to use html to build a frontend
from google.colab import output  # This is Necessary for JS-to-Python communication
```

Now, I have already defined the game state for you, all you need to do is to use this variable to complete the game logic. First, read the following code, especially the class attribute of `MemoryGame`.

```python
class MemoryGame:
    def __init__(self):
        self.animal_pool = ['🐶 Dog', '🐱 Cat', '🐭 Mouse', '🐹 Hamster', 
                            '🐰 Bunny', '🦊 Fox', '🐻 Bear', '🐼 Panda']
        self.cards = self.animal_pool * 2
        random.shuffle(self.cards)
        
        self.flipped_indices = []
        self.matched_indices = []
        self.moves = 0
        self.pairs_found = 0

    def reset(self):
        random.shuffle(self.cards)
        self.flipped_indices = []
        self.matched_indices = []
        self.moves = 0
        self.pairs_found = 0
```

Next is your job, complete the game logic below!

```python
# Initialize the backend instance
game = MemoryGame()

def handle_card_click(index):
    """This function is called by JavaScript whenever a card is clicked."""
    index = int(index)
    
    # Logic: Ignore if already matched or already flipped
    if index in game.matched_indices or index in game.flipped_indices:
        return

    # Add to currently flipped
    game.flipped_indices.append(index)
    
    # Tell JS to flip the card visually
    output.eval_js(f'uiFlipCard({index}, "{game.cards[index]}")')

    # Check if we have a pair
    if len(game.flipped_indices) == 2:
        game.moves += 1
        idx1, idx2 = game.flipped_indices
        
        if game.cards[idx1] == game.cards[idx2]:
            # It's a MATCH
            game.matched_indices.extend([idx1, idx2])
            game.pairs_found += 1
            game.flipped_indices = []
            output.eval_js(f'uiMarkMatch({idx1}, {idx2}, {game.moves}, {game.pairs_found})')
        else:
            # Not a match - tell JS to flip them back after a delay
            game.flipped_indices = []
            output.eval_js(f'uiFlipBack({idx1}, {idx2}, {game.moves})')

# Register the Python function so JavaScript can see it
output.register_callback('notebook.handle_card_click', handle_card_click)
```

The followings are some HTML, CSS and Javascript to make the game looks better. Since this is not a HTML or JS workshop, we won't talk about this part, and you also don't need to understand what each line does.

```python
html_template = f"""
<style>
    .board {{ display: grid; grid-template-columns: repeat(4, 1fr); gap: 10px; max-width: 400px; margin: 20px auto; }}
    .card {{ aspect-ratio: 1; background: #667eea; border-radius: 8px; display: flex; 
             align-items: center; justify-content: center; font-size: 1.5em; cursor: pointer; color: transparent; transition: 0.3s; }}
    .card.flipped {{ background: white; color: black; border: 2px solid #667eea; cursor: default; }}
    .card.matched {{ background: #4caf50; color: white; opacity: 0.6; cursor: default; }}
    .stats {{ text-align: center; font-family: sans-serif; }}
</style>

<div class="stats">
    <h3>Moves: <span id="moves">0</span> | Pairs: <span id="pairs">0</span>/8</h3>
</div>
<div class="board" id="board"></div>

<script>
    // Create the board visually
    const board = document.getElementById('board');
    for (let i = 0; i < 16; i++) {{
        const card = document.createElement('div');
        card.className = 'card';
        card.id = 'card-' + i;
        card.onclick = () => google.colab.kernel.invokeFunction('notebook.handle_card_click', [i], {{}});
        board.appendChild(card);
    }}

    // UI Update functions called by Python
    function uiFlipCard(index, content) {{
        const el = document.getElementById('card-' + index);
        el.classList.add('flipped');
        el.textContent = content;
    }}

    function uiMarkMatch(id1, id2, moves, pairs) {{
        document.getElementById('moves').textContent = moves;
        document.getElementById('pairs').textContent = pairs;
        [id1, id2].forEach(id => {{
            const el = document.getElementById('card-' + id);
            el.classList.remove('flipped');
            el.classList.add('matched');
        }});
        if(pairs === 8) alert("You won in " + moves + " moves!");
    }}

    function uiFlipBack(id1, id2, moves) {{
        document.getElementById('moves').textContent = moves;
        setTimeout(() => {{
            [id1, id2].forEach(id => {{
                const el = document.getElementById('card-' + id);
                el.classList.remove('flipped');
                el.textContent = "";
            }});
        }}, 1000);
    }}
</script>
"""
```

So, that's it. Easy, right? Let's go and play it!
