````python
# matplotlib.pyplot: Used for creating plots and charts (graphs, histograms, etc.), our protagonist
import matplotlib.pyplot as plt

# pandas: Used for reading and manipulating data (like CSV files, tables)
import pandas as pd

# numpy: Used for numerical operations (arrays, math functions, random numbers)
import numpy as np

# scipy.stats: Statistical functions
from scipy.stats import norm, gaussian_kde
from scipy import optimize

# norm: Normal (Gaussian) distribution functions
# gaussian_kde: Kernel Density Estimation - creates smooth curve from data
# optimize: Curve fitting functions (finding best-fit parameters)
````

"Okay," Chiaki said patiently, "now read the file."

````python
grades = pd.read_csv('final_grades.csv')
````

````python
# Create a new figure with specific dimensions
# figsize=(10,6) sets the width to 10 inches and height to 6 inches
plt.figure(figsize=(10, 6))

# Create a histogram (bar chart showing frequency distribution)
# grades["Grade"]: the data to plot (our list/array of student grades)
# bins=20: divides the grade range into 20 equal intervals/bars
# edgecolor='black': adds black borders around each bar for better visibility
# alpha=0.7: sets transparency to 70% (0 = fully transparent, 1 = fully opaque)
plt.hist(grades["Grade"], bins=20, edgecolor="black", alpha=0.7)

# Add a title to the plot
# The title appears above the graph
plt.title("Distribution of Final Grades - SSTIA1234J")

# Label the x-axis (horizontal axis)
# This tells viewers what the horizontal values represent (grades)
plt.xlabel("Grade")

# Label the y-axis (vertical axis)
# This tells viewers what the vertical values represent (count of students)
plt.ylabel("Number of Students")

# Add a grid to the plot for easier reading of values
# True: turns on the grid
# alpha=0.3: makes the grid lines very light (30% opacity) so they don't distract
plt.grid(True, alpha=0.3)

# Display the plot
# This actually shows the figure on screen
# Without this line, nothing would appear
plt.show()
````

"Also, we need to study how to make a curve fitting."

````python
# Create histogram
counts, bin_edges, _ = plt.hist(grades["Grade"], bins=20, alpha=0.7, edgecolor="black")
plt.close()  # Close the temporary plot

# Calculate bin centers (x-values for our fit)
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

# Define quadratic function
def quad(x, a, b, c):
    return a * x**2 + b * x + c

# Fit the quadratic to bin_centers (grades) vs counts (number of students)
params, _ = optimize.curve_fit(quad, bin_centers, counts)
a, b, c = params

# Generate smooth curve for plotting
x_smooth = np.linspace(min(bin_centers), max(bin_centers), 100)
y_smooth = quad(x_smooth, a, b, c)

# Figure size
plt.figure(figsize=(10, 6))

# Plot histogram
plt.hist(
    grades["Grade"],
    bins=20,
    edgecolor="black",
    alpha=0.7,
    label="Actual Grade Distribution",
)

# Plot quadratic fit
plt.plot(
    x_smooth,
    y_smooth,
    "r-",
    linewidth=2,
    label=f"Quadratic Fit: y = {a:.2f}x² + {b:.2f}x + {c:.2f}",
)

# Labels and title
plt.title("Distribution of Final Grades with Quadratic Fit - SSTIA1234J")
plt.xlabel("Grade")
plt.ylabel("Number of Students")
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
````

````python
# Now let's make a real grading map in GC, let's try!

# --- 1. Calculate Statistics ---
mean = np.mean(grades["Grade"])
std = np.std(grades["Grade"])
q1, q2, q3 = np.percentile(grades["Grade"], [25, 50, 75])

# --- 2. Initialize Plot ---
fig, ax = plt.subplots(figsize=(10, 8))

# --- 3. Plot Histogram ---
bin_width = 5
ax.hist(grades["Grade"], bins=range(0, 130, bin_width), color="white", edgecolor="black", linewidth=0.8)

# --- 4. Plot Distribution Curves ---
x = np.linspace(0, 130, 500)

scale_factor = len(grades["Grade"]) * bin_width

y_normal = norm.pdf(x, mean, std) * scale_factor
ax.plot(x, y_normal, color="blue", lw=1, label="Normal Distribution")

kde = gaussian_kde(grades["Grade"])
y_kde = kde(x) * scale_factor
ax.plot(x, y_kde, color="darkred", linestyle="--", lw=1, label="Estimated Distribution")

# --- 5. Add the Boxplot ---
ax.boxplot(
    grades["Grade"],
    vert=False,
    positions=[70.0],  # Set at the calculated height
    widths=10,  # Scale box thickness relative to histogram height
    patch_artist=True,
    showfliers=False,
    boxprops=dict(facecolor="#D3D3D3", color="black"),
    medianprops=dict(color="#ff7f0e"),
)

# --- 6. Add Stats Text ---
stats_text = (
    f"Q1: {q1:.1f}\n"
    f"Q2: {q2:.1f}\n"
    f"Q3: {q3:.1f}\n"
    f"Mean: {mean:.3f}\n"
    f"Std: {std:.3f}"
)
# --- 7. Final Aesthetics ---
ax.set_ylim(0, 95)
ax.set_yticks(np.arange(0, 80, 10))
ax.yaxis.set_major_formatter(plt.FormatStrFormatter("%.1f"))
ax.set_xlabel("Score", fontsize=12)
ax.set_ylabel("Frequency", fontsize=12)
ax.set_xlim(0, 130)
ax.set_ylim(0, 90)

ax.legend(loc="upper right", frameon=True, borderpad=1)
plt.title("SSTIA1234J 26FA Final Exam Grade Distribution", fontsize=14, pad=20)

plt.show()
````

Now, Let's make a quick recitation about what we learn:

### 1. The Two Ways to Plot

* **Pyplot State-machine (Simple):** Uses `plt.plot()`. Good for quick, one-off charts.
* **Object-Oriented Style (Professional):** Uses `fig, ax = plt.subplots()`. This is what we used for your Grade Distribution because it allows you to overlay different plots (like a Boxplot on a Histogram) on the same "axes" object.

### 2. Basic Plotting Recipe

Every Matplotlib script follows these four steps:

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Prepare Data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# 2. Create Figure and Axes
fig, ax = plt.subplots(figsize=(8, 4))

# 3. Plot Data
ax.plot(x, y, label='Sine Wave', color='blue', linewidth=2)

# 4. Customize & Show
ax.set_title("My First Plot")
ax.set_xlabel("Time")
ax.set_ylabel("Amplitude")
ax.legend()
plt.show()
```

### 3. Essential Chart Types

| Function | Type | Best Usage |
| --- | --- | --- |
| `ax.plot()` | **Line** | Trends over time or continuous functions. |
| `ax.scatter()` | **Scatter** | Relationships/correlations between two variables. |
| `ax.bar()` | **Bar** | Comparing quantities across categories. |
| `ax.hist()` | **Histogram** | Showing the distribution/frequency of data. |
| `ax.boxplot()` | **Boxplot** | Visualizing medians, quartiles, and outliers. |

### 4. Saving Your Plot

To save a high-quality version of your exam distribution for a report:
