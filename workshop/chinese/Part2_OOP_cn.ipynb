### 2.1 **什么是 OOP？**

**面向对象编程 (Object-Oriented Programming, OOP)** 是一种将代码组织成可重用、自包含对象的编程范式。每个对象包含：
- **数据**（属性/特征）
- **行为**（方法/函数）

#### 核心概念：
- **类 (Class)**：创建对象的蓝图/模板
- **对象 (Object)**：类的实例
- **封装 (Encapsulation)**：将数据和方法捆绑在一起
- **继承 (Inheritance)**：基于现有类创建新类
- **多态 (Polymorphism)**：为不同的数据类型使用统一的接口
- **抽象 (Abstraction)**：隐藏复杂的实现细节


### 2.2 **类与对象**

#### 创建基础类：

你可以使用 `class` 关键字创建类。在类中，会包含属性（变量）和方法（函数）：

```python
class Dog:
    # Class attribute (shared by all instances)
    species = "Canis familiaris"
    
    # Constructor method (initializer)
    def __init__(self, name, age):
        # Instance attributes (unique to each instance)
        self.name = name
        self.age = age
    
    # Instance method
    def bark(self):
        return f"{self.name} says woof!"
    
    def get_info(self):
        return f"{self.name} is {self.age} years old"

# Creating objects (instances)
dog1 = Dog("Buddy", 3)
dog2 = Dog("Max", 5)

print(dog1.bark())  # "Buddy says woof!"
print(dog2.get_info())  # "Max is 5 years old"

# TODO: Instantiate a dog called Jerry, let it bark
```

`self` 是代表实例本身的变量。对于每个实例方法，必须将 `self` 作为**第一个**传递给函数的变量。

你可能已经注意到，像 `__init__` 这样带有双下划线的方法是内置函数，被称为“dunder methods"。当使用 `dog1 = Dog("Buddy", 3)` 创建实例时，它会隐式调用 `__init__(self, "Buddy", 3)`。还有很多dunder methods，我们稍后会详细讨论它们以及如何定义我们自己的dunder methods。

```python
print(dog1.__dict__)
print(dog1.__dir__())
print(dog1.__hash__())
dog1.__init__("Buddy2", 3)
print(dog1.name)
print(dog1.__sizeof__())
```

有趣的一点是，当调用 `dog1.__dir__()` 时，你在目录中看不到类属性 `species`。这揭示了 Python 中实例属性和类属性的区别：前者存储在每个唯一的实例内部，而后者存储在类中并由所有实例共享。


### 2.3 **OOP 的四大支柱**

#### 2.3.1 封装 (Encapsulation)
封装是将数据及操作该数据的方法捆绑在一个单元（类）中。你可以通过使用 `__` 前缀来控制属性和方法的可见性，例如 `__balance` 表示私有属性，或者 `__helper(self)` 表示私有辅助方法。更准确地说：`__` 前缀表示 **私有 (private)**（类外部不可见且不能被继承）；`_` 前缀表示 **受保护 (protected)**（类外部不可见但可以被继承）；无前缀则表示 **公有 (public)**（始终可见且可被继承）。

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.__balance = balance  # Private attribute
        
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            return f"Deposited ${amount}"
        return "Invalid amount"
    
    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            return f"Withdrew ${amount}"
        return "Insufficient funds"
    
    def get_balance(self):
        return self.__balance

account = BankAccount("John", 1000)
print(account.deposit(500))
print(account.get_balance())  # 1500

## Don't call the private attribute outside the class.
# account.__balance  # Error: Attribute is private
```


#### 2.3.2 继承 (Inheritance)
继承是基于现有类创建新类。派生类（子类）将继承基类（父类）中的公有和受保护的属性及方法。

```python
# Parent/Base class
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        return "Some sound"

# Child/Derived class
class Cat(Animal): # This line means Cat is derived form Animal 
    def __init__(self, name, color):
        super().__init__(name)  # Call parent's __init__
        self.color = color
    
    # Method overriding
    def speak(self):
        return "Meow!"
    
    def describe(self):
        return f"{self.name} is {self.color}"

# Another child class
class Dog(Animal):
    #TODO: define the __init__ and describe methods for Dog
    
    
    def speak(self):
        return "Woof!"

cat = Cat("Whiskers", "orange")
dog = Dog("Buddy", "White")

print(cat.speak())  # "Meow!"
print(dog.speak())  # "Woof!"

print(cat.__dir__())
print(dog.__dir__())

# TODO: Called the describe method for cat and dog
```

```python
### Types of Inheritance

# Single Inheritance
class Parent:
    pass

class Child(Parent):
    pass

# Multiple Inheritance
class Mother:
    def method1(self):
        return "From Mother"

class Father:
    def method2(self):
        return "From Father"

class Child(Mother, Father):
    pass

# Multi-level Inheritance
class GrandParent:
    pass

class Parent(GrandParent):
    pass

class Child(Parent):
    pass

# Hierarchical Inheritance
class Vehicle:
    pass

class Car(Vehicle):
    pass

class Bike(Vehicle):
    pass
```

#### 2.3.3 多态 (Polymorphism)
多态意味着“相同的接口，不同的实现”。它允许我们通过同一个接口调用不同派生类的方法。

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height

class Circle:
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return 3.14 * self.radius ** 2

# TODO: Add a class called Triangle to calculate the area of triangle

# Same method name, different implementations
shapes = [Rectangle(4, 5), Circle(3)]
for shape in shapes:
    print(f"Shape: {shape.__class__} Area: {shape.area()}")
```

#### 2.3.4 抽象 (Abstraction)
抽象意味着隐藏复杂的实现细节。在 Python 中，你可以使用 `@abstractmethod` 创建一个抽象基类，以此来运用多态。

```python
from abc import ABC, abstractmethod

class Shape(ABC):  # Abstract class
    @abstractmethod
    def area(self):
        pass
    
    @abstractmethod
    def perimeter(self):
        pass

class Square(Shape):
    def __init__(self, side):
        self.side = side
    
    def area(self):
        return self.side ** 2
    
    def perimeter(self):
        return 4 * self.side

# shape = Shape()  # Error: Cannot instantiate abstract class
square = Square(5)
print(square.area())  # 25
```

### 2.4 **特殊方法（双下划线方法）**

这些是以双下划线开头和结尾的方法，定义了特定的行为。在之前的章节中我们讨论了一些基础的dunder methods，但实际上，我们可以重写这些方法来创建自定义行为。例如，我们可以通过重写 `__eq__`、`__lt__` 和 `__le__` 来分别定义“等于”、“小于”和“小于等于”的比较逻辑。

```python
class Book:
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages
    
    # String representation
    def __str__(self):
        return f"{self.title} by {self.author}"
    
    def __repr__(self):
        return f"Book('{self.title}', '{self.author}', {self.pages})"
    
    # Length
    def __len__(self):
        return self.pages
    
    # Comparison
    def __eq__(self, other):
        return self.title == other.title and self.author == other.author
    
    # Less than
    def __lt__(self, other):
        return self.pages < other.pages
    
    # TODO: define <= sign by __le__

book = Book("Python 101", "John Doe", 300)
book2 = Book("C++ 151", "Asya Shubina", 3000)
print(str(book))   # "Python 101 by John Doe"
print(repr(book2))
print(len(book))   # 300
print(book < book2)     # True
```

### 2.5 **类方法与静态方法**

到目前为止，所有的方法都需要将 `self` 作为第一个参数，这种方法被称为实例方法。事实上，还有类方法和静态方法。类方法需要将 `cls` 作为第一个参数来获取类属性，并添加 `@classmethod` 装饰器；静态方法既不需要 `self` 也不需要 `cls`，只需添加 `@staticmethod` 装饰器，这意味着它不能修改实例属性或类属性。

```python
class Pizza:
    # Class attribute
    restaurant = "Mario's Pizza"
    
    def __init__(self, toppings):
        self.toppings = toppings
    
    # Instance method - works with instance data
    def description(self):
        return f"Pizza with {', '.join(self.toppings)}"
    
    # Class method - works with class, not instance
    @classmethod
    def change_restaurant(cls, new_name):
        cls.restaurant = new_name
    
    @classmethod
    def margherita(cls):
        return cls(["tomato", "mozzarella", "basil"])
    
    # Static method - doesn't access class or instance
    @staticmethod
    def calculate_area(radius):
        return 3.14 * radius ** 2

# Using class method
pizza1 = Pizza.margherita()
Pizza.change_restaurant("Luigi's Pizza")

# Using static method
area = Pizza.calculate_area(10)
```


### 2.6 **属性装饰器 (Getters/Setters)**

在前面的章节中，我们提到了可见性。实际上，为了实现更好的封装，我们通常希望属性是私有或受保护的。在这种情况下，我们需要使用 Getter 和 Setter 作为属性装饰器来读取或修改数据。

```python
class Person:
    def __init__(self, name, age):
        self._name = name
        self._age = age
    
    @property
    def name(self):
        return self._name
    
    @name.setter
    def name(self, value):
        if isinstance(value, str) and len(value) > 0:
            self._name = value
        else:
            raise ValueError("Name must be a non-empty string")
    
    @property
    def age(self):
        return self._age
    
    @age.setter
    def age(self, value):
        if 0 <= value <= 120:
            self._age = value
        else:
            raise ValueError("Age must be between 0 and 120")

person = Person("Alice", 30)
print(person.name)  # Getter: "Alice"
person.name = "Bob"  # Setter
# person.age = 150  # Raises ValueError
```

### 2.7 **一切皆对象** 

在 Python 中，所有的值（从某种意义上说）都是对象。

因此，`356, 3.2, 4+5j, "Hello World", False` 都是对象。

不相信吗？试试这些代码：

```python
print(isinstance(2020, object)) 
print(isinstance(4.33, object))
print(isinstance(5j, object))
print(isinstance(True, object))
print(isinstance("COVID-19", object))
```


所以，以下内容实际上是针对相应内置类型的对象构造函数。它们分别接收一些输入，并根据输入创建该类型的实例（如果输入有效）。尝试运行以下代码（并留意默认值）。

```python
int()
float()
complex()
bool()
str()
int(-9.789)
float(9)
float("nan")
float("inf")
float("-inf")
complex(1,2)
bool(1)
bool("")
bool(5)
str("Hello")
str(10)
```

更令人惊讶的是，甚至模块也是对象。

```python
import math

print(type(math))  # <class 'module'>
print(math.__name__)  # "math"

print(isinstance(math, object))
```

如果你是从其他编程语言转过来的，你可能会惊讶地发现，在 Python 中你可以轻松地“改变”变量的数据类型。例如，这段代码在 Python 中完全有效，但在 C++ 中则毫无道理。

```python
num = 1.00
print(type(num))
num = 2+1j
print(type(num))
```

为了理解这一点，我们需要知道 Python 中的变量更像是**指针**。它们指向堆（动态内存分配）中的某个对象。

如果你学过 C/C++，指针是一个变量，其值是它所指向的内容的地址。指针变量也被静态绑定为某种特定类型。在我们的讨论中，只需将指针视为指向内存堆中的一个位置。
